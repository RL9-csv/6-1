```python
def hour_func(x):
    if (x >= 6) & (x < 12):
        return 'morning'
    elif (x >= 12) & (x < 18):
        return 'afternoon'
    elif (x >= 18) & (x < 24):
        return 'night'
    else:
        return 'evening'
```

```python
cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)
```

```python
cc_df
```

```python
cc_df.groupby(['hour_cat']).count()
```
